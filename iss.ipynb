```python
import requests
import matplotlib.pyplot as plt
import time

x_plot = []
y_plot = []

plt.title("ISS Position")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xlim(-180, 180)
plt.ylim(-90, 90)

for i in range(100):
    response = requests.get('https://api.wheretheiss.at/v1/satellites/25544', timeout=10)
    if response.status_code == 200:
        data = response.json()
        latitude = data.get('latitude')
        longitude = data.get('longitude')
        print(f"{i+1}. Latitude: {latitude}, Longitude: {longitude}")
        x_plot.append(longitude)
        y_plot.append(latitude)
    else:
        print(f"{i+1}. Request failed with status code: {response.status_code}")
    time.sleep(10)

plt.plot(x_plot, y_plot, marker='o', linestyle='-')
plt.show()
```
